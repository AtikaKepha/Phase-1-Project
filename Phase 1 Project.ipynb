```python
#Extract tn movie budget data
df4 = pd.read_csv('tn.movie_budgets.csv')
```

```python
df4.head()
```

```python
#Strip dollar signs from revenue data
df4.production_budget = df4.production_budget.apply(lambda x: x.strip('$'))
df4.domestic_gross = df4.domestic_gross.apply(lambda x: x.strip('$'))
df4.worldwide_gross = df4.worldwide_gross.apply(lambda x: x.strip('$'))
```

```python
#remove commas from revenue data
df4.production_budget = df4.production_budget.apply(lambda x: x.replace(',', ''))
df4.domestic_gross = df4.domestic_gross.apply(lambda x: x.replace(',',''))
df4.worldwide_gross = df4.worldwide_gross.apply(lambda x: x.replace(',',''))
```

```python
df4.head()
```

```python
df4.info()
```

```python
df4.describe()
```

```python
#convert production budget and movie revenue data to numeric data
df4["production_budget"] = pd.to_numeric(df4["production_budget"])
df4["domestic_gross"] = pd.to_numeric(df4["domestic_gross"])
df4["worldwide_gross"] = pd.to_numeric(df4["worldwide_gross"])
```

```python
df4.info()
```

```python
#summary statistics
df4.describe()
```

```python
#sort worldwide movie revenue in descending order, highest values first.
top_gross = df4.sort_values('worldwide_gross',ascending=False)
top_gross.head(10)
```

```python
#New dataframe containing movie, production and revenue data
top_gross_2 = top_gross[['movie', 'production_budget','worldwide_gross']]
```
